## Series.

### Tworzenie obiektu, indeksowanie, wstawianie brakujących wartości, daty jako indeks.

Aby stworzyć serię danych, korzystamy z funkcji **Series**. Aby zdefiniować, czym jest obiekt typu Series - patrz punkt 2 wprowadzenia.

```python
s = pd.Series([9, 1.2, 6, 0])
print(s)
```

Funkcja Series przypisuje automatycznie indeks serii danych - począwszy oczywiście od 0. Możemy jednak podać własne wartości indeksowania, za pomocą parametru index.

```python
s1 = pd.Series([2, 9, 1, 4], index=['a', 'b', 'c', 'd'])
print(s1)
```

Po takich indeksach możemy również śmiało wycinać. Co ważne, również w serii typ danych jest jednorodny.

```python
s2 = pd.Series([2, 9, 1, '4'], name='Seryjka')
print(s2)
```

Typ object to oczywiście odpowiednik pythonowego stringa. Możemy nadać naszej serii nazwę przy pomocy parametru name.

Konwersja typów int na typy float jest podobna do rozwiązań zastosowanych w NumPy.

```python
s3 = pd.Series([2, 9, 1, 4], dtype=float)
print(s3)
```

```python
s3 = pd.Series([2, 9, 1, 4.])
print(s3)
```

Jak wstawić braki w danych? Musimy skorzystać z atrybutu **nan** w bibliotece NumPy.

```python
import numpy as np
s4 = pd.Series([2, 9, np.nan, 4])
print(s4)
```

Jest to istotne o tyle, że w codziennej pracy regularnie możemy się spotkać z brakami w danych. Jeżeli mamy do czynienia z brakiem, to np. przy liczeniu średniej dla serii wartość NaN nie jest brana pod uwagę.

```python
s4.mean()
```

Gdybyśmy zamiast NaN mieli w serii 0, wpłynęłoby to na średnią.

```python
s4[2] = 0
print(s4)
print(s4.mean())
```

Możemy oczywiście tworzyć obiekty, w skład których wchodzą inne typy danych - boolowskie, liczby zespolone.

```python
s = pd.Series([True, False, True])
print(s)
```

```python
s1 = pd.Series([1 + 2j, 3 + 4j, 5 + 6j])
print(s1)
```

Bardzo użytecznym jest - w razie pracy z seriami czasowymi - jest użycie jako indeksu dat. Podajemy w takim wypadku jako wartość parametru index wartość funkcji **date_range** - wskazujemy jako parametry dzień początkowy i liczbę dni ze wskazanym włącznie, które użyjemy jako indeks.

```python
s2 = pd.Series(np.arange(10, 20), index=pd.date_range('20240801', periods=len(np.arange(10, 20))))
print(s2)
```

Oprócz typu danych mamy również podaną częstotliwość czasową, tj. jakie są czasowe różnice między indeksami.

Zmienić indeks możemy również w istniejącym obiekcie Series.

```python
s.index = pd.date_range('20230101', periods=len(s))
print(s)
```

```python
s1.index = np.arange(1, len(s1) + 1)
print(s1)
```

### Metoda apply.

Metoda ta pozwala na zastosowanie do serii (jak i do DataFrame'a) funkcji bardziej niestandardowych, zbudowanych wg naszego upodobania.

```python
price
```

# Case study - dane giełdowe.

Po wstępnych uwagach dot. obiektów Pandas pora na pracę z realnymi danymi, bo przecież to jest główny cel tej biblioteki. Zdefiniujemy funkcję służącą pobraniu danych giełdowych z serwisu stooq.pl. Domyślnie jako argument wskazany jest ticker spółki Amazon. Jeżeli chcemy pobrać dane dot. innej spółki, powinniśmy wskazać alternatywny ticker. Importujemy bibliotekę pandas-datareader, która pomoże nam pobrać dane i zapisać je w formie obiektu dataframe.

## Ładowanie danych.

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')
```

Oprócz utworzenia obiektu podejrzymy jego pierwszych pięć pierwszych rekordów (z pomocą funkcji **head**, która domyślnie wyświetla pięć wierszy od góry, można przy zadeklarować inną liczbę).

```python
df = fetch_financial_data()
df.head()
```

## Wstępne zapoznanie się z obiektem.

Funkcja została wywołana i stworzyła ona obiekt z 1256 wierszami, 5 kolumnami (kolejno cena otwarcia, najwyższa, najniższa, zamknięcia i wolumen) i indeksami w formie dat z ostatnich 5 lat.

```python
df.info()
```

Istotne dla analityka giełdowego - najważniejszą ceną jest cena zamknięcia.

Jeżeli z kolei chcemy podejrzeć ostatnie wiersze, użyjmy metody **tail** (domyślnie również wyświetla 5 wierszy, tym razem wskażemy liczbę 14 dla podejrzenia ostatnich 14 wierszy).

```python
df.tail(14)
```

Podejrzyjmy nazwy kolumn.

```python
df.columns
```

Możemy zamienić występujące w nich litery na małe. W tym celu zastosujemy wyrażenie listowe, modyfikujący każdy element listy metodą **lower**.

```python
df.columns = [col.lower() for col in df.columns]
df.columns
```

## Obliczanie nowych kolumn.

Kolejne bardzo częste zagadnienie to obliczanie nowych kolumn. Ponownie załadujemy dane za pomocą zdefiniowanej wcześniej funkcji (tym razem dla spółki Microsoft) i standardowo wyciągniemy kilka podstawowych informacji nt. obiektu dataframe.

```python
df = fetch_financial_data(company='MSFT')
df
```

```python
df.info()
```

```python
df.describe()
```

Teraz będziemy chcieli zbudować nową kolumnę, która pokaże średnią cenę z cen otwarcia i zamknięcia. Najprostszym sposobem jest dodanie obu kolumn i podzielenie ich przez 2.

```python
df['Average'] = (df.Open + df.Close) / 2
df
```

W kolejnej kolumnie chcemy mieć średnią zmianę procentową. Najpierw jednak posortujmy obiekt, żeby daty najwcześniejsze były na górze (czyli odwrotnie do obecnego obiektu). Zastosujemy metodę **sort_index** - możemy przypisać posortowany obiekt do zmiennej, możemy też - dzięki parametrowi inplace o wartości True - posortować go w miejscu, modyfikując obiekt.

```python
df.sort_index(ascending=True, inplace=True)
df
```

Przystąpmy do stworzenia kolumny ze średnią zmianą procentową. Aby ją policzyć, będziemy badać cenę zamknięcia z sesji na sesję. Za odwołanie się do poprzedniej wartości odpowiada metoda **shift**, przyjmująca jako parametr interwał czasowy (może być negatywna - odwołanie do następnego dnia). Z tego powodu dla pierwszej pozycji został ustawiony NaN - brak wartości, do której można się odwołać.

```python
df['Close_shift'] = df.Close.shift(1)
df
```

Mając odwołanie do poprzedniej wartości, możemy łatwo policzyć tę zmianę zmiany. Dzielimy wartość zamknięcia przez wartość zamknięcia dzień przed i odejmujemy 1.

```python
df['Daily_change'] = (df.Close / df.Close_shift) - 1
df
```

Jeżeli chcemy usunąć niepotrzebną kolumnę, możemy użyć słowa kluczowego **del** i nazwy kolumny, ewentualnie metody **drop** (z przypisaniem do zmiennej lub parametrem inplace=True).

```python
del df['Close_shift']
df
```

Teraz wyciągnijmy z utworzonej kolumny wartość minimalną i maksymalną metodami min i max.

```python
print(df['Daily_change'].min())
print(df['Daily_change'].max())
```

Na sam koniec wyświetlmy histogram kolumny "Daily_change". Za wyświetlanie wykresu tego typu z poziomu Pandas odpowiada metoda **hist** - możemy wskazać m.in. parametr bins, czyli przedziały, w jakich mieszczą się poszczególne słupki.

```python
df['Daily_change'].hist(bins=100)
```

Wyświetlimy również cenę zamknięcia w formie wykresu liniowego - metoda **plot**.

```python
df.Close.plot()
```

## Filtrowanie danych.

Często będą nas interesować określone fragmenty naszego obiektu, dlatego umiejętność ich filtracji jest szalenie istotna.

Wytnijmy np. te dni sesyjne, w których dzienna stopa zwrotu była wyższa od 0. Właściwą składnią jest wpisanie wewnątrz nawiasów kwadratowych po nazwie zmiennej warunku logicznego.

```python
df[df.Daily_change > 0]
```

Sam warunek logiczny stworzy tylko maskę logiczną na nasze dane.

```python
df.Daily_change > 0
```

Możemy użyć też metody loc.

```python
df.loc[df.Daily_change > 0]
```

Łatwo możemy dzięki temu policzyć średnią dodatnią i ujemną stopę zwrotu.

```python
df[df.Daily_change > 0].Daily_change.mean()
```

```python
df[df.Daily_change <= 0].Daily_change.mean()
```

Sprawdźmy, czy jest szansa, żeby cena zamknięcia była ceną najwyższą.

```python
df[df.Close == df.High]
```

Mieliśmy trzy takie sytuacje na przestrzeni ostatnich 5 lat.

```python
df[df.Close == df.Low]
```

Ani razu nie zdarzyła się za to sytuacja, że cena zamknięcia była najniższą.

Chcemy wycinać też dane dot. konkretnych notowań. Dobierzmy się do indeksu.

```python
df.index
```

Wybierzmy sesje po 5 sierpnia 2024 r.

```python
df[df.index > '2024-08-05']
```

A jak łączyć maski logiczne? Operatorami logicznymi. Wybierzmy np. wszystkie notowania od początku do końca grudnia 2022 r. - maski musimy wtedy otoczyć nawiasem okrągłym i połączyć jest symbolem "&" (ampersandem).

```python
df[(df.index >= '2022-12-01') & (df.index < '2023-01-01')]
```

Uwaga - nie możemy stosować słowa kluczowego and jako operatora logicznego - wywołanie takiego kodu zwróci błąd.

Operatorem alternatywy jest "|" (pipe). W tym wypadku wyświetlamy wszystkie rekordy przed 1 grudnia 2019 r. i po 1 stycznia 2023 r.

```python
df[(df.index < '2019-12-01') | (df.index > '2023-01-01')]
```

A jak możemy wyciąć dane dla danego miesiąca z konkretnego roku? Używamy atrybutów **month** i **year**.

```python
df[(df.index.month == 12) & (df.index.year == 2022)]
```

Możemy też wyfiltrować rekordy dla pierwszego dnia każdego miesiąca z okresu objętego obiektem.

```python
df[df.index.day == 1]
```

No i na koniec tego notatnika przedstawmy wykres liniowy zmiany ceny zamknięcia. Ustawimy przy tym styl wykresu przy pomocy biblioteki **Seaborn**, o której szerzej w dalszej części szkolenia.

```python
import seaborn as sns
sns.set_style('darkgrid')

df[df.index.day == 1].Close.plot()
```
